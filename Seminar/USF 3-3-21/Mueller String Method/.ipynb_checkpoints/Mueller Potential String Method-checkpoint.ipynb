```python
import numpy as np
import numpy.matlib as ml
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def Mueller(x,y):
    A=[-200,-100,-170,15]
    a=[-1,-1,-6.5,0.7]
    b=[0,0,11,0.6]
    c=[-10,-10,-6.5,0.7]
    x0=[1,0,-0.5,-1]
    y0=[0,0.5,1.5,1]
    result = 0.0
    for i in range(len(A)):
        result += A[i]*np.exp(a[i]*(x-x0[i])**2+b[i]*(x-x0[i])*(y-y0[i])+c[i]*(y-y0[i])**2)
    return result

def gMueller(x,y):
    A=[-200,-100,-170,15]
    a=[-1,-1,-6.5,0.7]
    b=[0,0,11,0.6]
    c=[-10,-10,-6.5,0.7]
    x0=[1,0,-0.5,-1]
    y0=[0,0.5,1.5,1]
    gx = 0.0
    gy = 0.0
    for i in range(len(A)):
        m = A[i]*np.exp(a[i]*(x-x0[i])**2+b[i]*(x-x0[i])*(y-y0[i])+c[i]*(y-y0[i])**2)
        gx+=m*(2*a[i]*(x-x0[i])+b[i]*(y-y0[i]))
        gy+=m*(b[i]*(x-x0[i])+2*c[i]*(y-y0[i]))
    return gx,gy

def getline(x0,y0,x1,y1,n):
    x=np.linspace(x0,x1,n)
    y=np.linspace(y0,y1,n)
    return x,y

def smooth(sx,sy,s):
    nx=sx[:]
    ny=sy[:]
    for i in range(1,len(nx)-2):
        nx[i] = (1-s)*sx[i] + s/2*(sx[i-1]+sx[i+1])
        ny[i] = (1-s)*sy[i] + s/2*(sy[i-1]+sy[i+1])
    return nx,ny

def reparameterize(sx,sy):
    n=len(sx)
    L=np.zeros(n)
    s=np.zeros(n)
    for i in range(1,n):
        dz = np.sqrt((sx[i]-sx[i-1])**2 + (sy[i]-sy[i-1])**2)
        L[i] = L[i-1] + dz
#        print('{:d} {:.6f} {:.6f}'.format(i,dz,L[i]))
    dl=L[n-1]/(n-1)
    nx=sx[:]
    ny=sy[:]
#    print('intervals: {:d}'.format(n-1))
#    print('contour length in: {:.6f}'.format(L[n-1]))
#    print('increment: {:.6f}'.format(dl))
    for m in range(1,n-1):
        s=m*dl
        dz=L[m]-L[m-1]
        f=(s-L[m-1])/dz
        nx[m] = sx[m-1] + f*(sx[m]-sx[m-1])
        ny[m] = sy[m-1] + f*(sy[m]-sy[m-1])
        ndz=np.sqrt((nx[m]-nx[m-1])**2 + (ny[m]-ny[m-1])**2)
#        print('{:d} {:.6f} {:.6f} {:.6f} {:.6f} {:.6f}'.format(m,s,L[m-1],dz,f,ndz))    
    return nx,ny

def write_string(fn,sx,sy):
    with open(fn,'w') as f:
        for i,(x,y) in enumerate(zip(sx,sy)):
            f.write('{:d} {:.6f} {:.6f}\n'.format(i,x,y))
        print('Wrote {}.'.format(fn))
def read_string(fn):
    with open(fn,'r') as f:
        sx=[]
        sy=[]
        for l in f:
            dat=l.split()
            sx.append(float(dat[1]))
            sy.append(float(dat[2]))
    return sx,sy
```

```python
nx=200
ny=200
x=np.linspace(-1.5,1.25,nx)
y=np.linspace(-0.25,2.0,ny)
z=ml.zeros((nx,ny))
xv, yv = np.meshgrid(x, y, sparse=False, indexing='xy')
zmin=1.e9
for i in range(nx):
    for j in range(ny):
        z[i,j]=Mueller(xv[i,j],yv[i,j])
        if z[i,j] < zmin:
            zmin=z[i,j]
for i in range(nx):
    for j in range(ny):
        z[i,j]-=zmin
```

# STRING METHOD

```python
nimg=24
sx,sy=getline(0.25,1.75,1.0,0.25,nimg)
niters=[1000]*7
stepsizes=[0.00001,0.00002,0.00004,0.00006,0.00006,0.00006,0.00006]
smos=[0.1]*4+[0.075]*2+[0.06]

ntotiter=0
write_string('{:05d}-string.dat'.format(ntotiter),sx,sy)
for stg,(niter,stepsize,smo) in enumerate(zip(niters,stepsizes,smos)):
    for i in range(niter):
        for j in range(nimg):
            gx,gy=gMueller(sx[j],sy[j])
            sx[j]-=stepsize*gx
            sy[j]-=stepsize*gy
        sx,sy=smooth(sx,sy,smo)
        sx,sy=reparameterize(sx,sy)
        if ntotiter>0:
            if stg==0 and ntotiter%10==0:
                write_string('{:05d}-string.dat'.format(ntotiter),sx,sy)
            elif ntotiter%100 == 0:
                write_string('{:05d}-string.dat'.format(ntotiter),sx,sy)
        ntotiter+=1
#plt.plot(sx,sy,linewidth=1.5,marker='o',color=scol((ntotiter+1)/(sum(niters)+1)))
write_string('{:05d}-string.dat'.format(ntotiter),sx,sy)
```

# CLIMBING STRING

```python
nimg=12
sx,sy=getline(-0.558561, 1.441388,-0.4,1.25,nimg)
oldx=sx.copy()
oldy=sy.copy()
niters=[1000,10000,10000]
stepsizes=[0.0000001,0.000001,0.000005]#,0.00002,0.00004,0.00006,0.00006,0.00006,0.00006]
smos=[0.1]*3#*4+[0.075]*2+[0.06]
nu=2
ntotiter=0
write_string('{:05d}-cstring.dat'.format(ntotiter),sx,sy)
for stg,(niter,stepsize,smo) in enumerate(zip(niters,stepsizes,smos)):
    for i in range(niter):
        for j in range(nimg):
            gx,gy=gMueller(sx[j],sy[j])
            oldx[j] = sx[j]
            oldy[j] = sy[j]
            sx[j] = oldx[j]-stepsize*gx
            sy[j] = oldy[j]-stepsize*gy
#            print(j,sx[j],tmp,sx[j]-tmp,stepsize*gx)
#        print(j,gx,gy)
        # local tangent unit vector
        dx=oldx[nimg-2]-oldx[nimg-3]
        dy=oldy[nimg-2]-oldy[nimg-3]
        al=np.sqrt(dx*dx+dy*dy)
        dx/=al
        dy/=al
        upx=sx[nimg-1]-oldx[nimg-1]
        upy=sy[nimg-1]-oldy[nimg-1]
        dot=upx*dx+upy*dy
        dx*=-dot*nu
        dy*=-dot*nu
        #print(dx,dy)
        sx[nimg-1]+=dx
        sy[nimg-1]+=dy
#        print(upx,dot,dx)
        if i%10==0:
            sx,sy=smooth(sx,sy,smo)
            sx,sy=reparameterize(sx,sy)
        if ntotiter>0:
            if stg==0 and ntotiter%100==0:
                write_string('{:05d}-cstring.dat'.format(ntotiter),sx,sy)
            elif stg==1 and i<1000 and ntotiter%100 == 0:
                write_string('{:05d}-cstring.dat'.format(ntotiter),sx,sy)
            elif ntotiter%100 == 0:
                write_string('{:05d}-cstring.dat'.format(ntotiter),sx,sy)
        ntotiter+=1
#plt.plot(sx,sy,linewidth=1.5,marker='o',color=scol((ntotiter+1)/(sum(niters)+1)))
write_string('{:05d}-cstring.dat'.format(ntotiter),sx,sy)
```
